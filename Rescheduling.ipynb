```python
import os
import pandas as pd
from sklearn.metrics.pairwise import linear_kernel
import random
from datetime import datetime, timedelta
df = pd.read_csv(os.getcwd()+"/Prediction Data.csv")
pred_mood = df.copy()

#Set available time for work
available_time=[[14,18],[10,13,14,17],[15,18,19,22],[11.5,12,18.5,22]] #[[day1 begin time, end time, begin time, end time],[day2],[day3]..]

column_list = ["Title", "Description", "Importance", "Difficulty", "Expected hour", "Expected minute", "Deadline"] #Deadline: mm/dd/yyyy

Assignment1 = ["w1","None",5,5,3,15,"01/31/2022"]
Assignment2 = ["w2","None",2,1,1,30,"01/30/2022"]
Assignment3 = ["w3","None",5,1,2,15,"01/28/2022"]
Assignment4 = ["w4","None",1,5,4,45,"01/29/2022"]

work = pd.DataFrame([Assignment1,Assignment2,Assignment3,Assignment4], columns=column_list)

#Expected time for each work converted to hours
for i in range(len(work["Title"])):
    if work["Expected minute"][i] < 31:
        work["Expected hour"][i]+=.5
    else:
        work["Expected hour"][i]+=1

#Predicted mood data taken from Panha's data
today = datetime.now() + timedelta(1) 
j=0
for i in range(len(df["timestamp"])):
    d, t = df["timestamp"][i].split()
    if t=='0:00':
        j+=1
    d = today + timedelta(j)
    pred_mood["timestamp"][i] = d.strftime('%m-%d-%Y') + " " + t

#Create the available timeslot from available time for work and mood index from predicted mood data
column_list2 = ["timestamp", "Title", "Description", "Importance", "Difficulty", "Expected hour", "Expected minute", "Stress&Worries", "Chaotic", "Happiness","Energy","Focus"]
avai_timeslot=pd.DataFrame([], columns=column_list2)

for i in range(len(available_time)):
    k=0
    while k < len(available_time[i]):
        begin = available_time[i][k]
        end = available_time[i][k+1]
        
        while begin < end + 1:
            if begin%1 == 0:
                m = "00"
            else:
                m = "30"
            t1, t2 = str(begin//1 + .00).split(".")
            t = str(t1) + ":" + m
            d = today + timedelta(i)
            slot = d.strftime('%m-%d-%Y') + " " + t
            for l in range(len(pred_mood["timestamp"])):
                if slot == pred_mood["timestamp"][l]:
                    stress = pred_mood["Stress&Worries"][l]
                    chaotic = pred_mood["Chaotic"][l]
                    happiness = pred_mood["Happiness"][l]
                    energy = pred_mood["Energy"][l]
                    focus = pred_mood["Focus"][l]
                    break
                else:
                    stress = ""
                    chaotic = ""
                    happiness = ""
                    energy = ""
                    focus = ""
            
            new_slot=pd.DataFrame([[slot,"","","","","","",stress,chaotic,happiness,energy,focus]], columns=column_list2)
            avai_timeslot=pd.concat([avai_timeslot, new_slot], ignore_index=True)
            
            begin += 0.5
        k+=2

```
